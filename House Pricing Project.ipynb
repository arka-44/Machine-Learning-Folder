```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.datasets
from sklearn import metrics
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
```

```python
hp_data = pd.read_csv("housing.csv")
hp_data
```

```python
hp_data.describe()
```

```python
X = hp_data.drop(columns="price" , axis = 1)
Y = hp_data["price"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["mainroad" , "guestroom" , "basement" , "hotwaterheating" , "airconditioning" ,"prefarea" ,"furnishingstatus"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot" , one_hot , categorical_features)] , remainder="passthrough")
transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
pd.DataFrame(transformed_X)
```

```python
X_test , X_train , Y_test , Y_train = train_test_split(transformed_X , Y , test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(transformed_X , Y)
model.score(X_test , Y_test)
```

```python
from sklearn.metrics import r2_score
prediction = model.predict(X_test)
r2_score(y_true = Y_test , y_pred = prediction)
```
